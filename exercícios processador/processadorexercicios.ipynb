### 12) Considere um computador com 64K células de memória, instruções de um operando, tendo possibilidade de ter um conjunto de 256 instruções de máquina. Considerando que cada instrução tem o tamanho de uma célula, que é o mesmo tamanho da palavra do sistema. Qual o tamanho, em bits, do Reg, CI e RDM? Qual é o total de bits dessa memória?

A memória tem 64K células, ou seja, 2^16 células, então o código de operação terá 16 bits. Se temos 256 (2^8) instruções, o operando terá 8 bits.

Logo, uma instrução terá 24 bits (16 bits do código + 8 de operando), consequentememte, a célula, a palavra, o Reg./ACC e o RDM terão 24 bits.

A memória tem 16 bits de endereço, assim o CI e o RDM terão 16 bits.

O total de bits da memória é dado por número de células x tamanho da célula, então o total de bits será 2^16 x 24 

```python
import math 

celulas = 64 * 1024 
codigoop = math.log2(celulas) 
bitsend = codigoop
qtdinst = 256
operando = math.log2(qtdinst)
instrucao = codigoop + operando

totalbits = celulas * instrucao

print(f"O Reg/ACC e o RDM têm {instrucao} bits")
print(f"CI e o REM têm {bitsend} bits")
print(f"O total de bits na memória é {totalbits}")
```

### 14) Um computador possui um conjunto de 128 instruções de um operando; supondo que sua memória tenha capacidade de armazenar 512 palavras e que cada instrução tenha o tamanho de uma palavra e da célula de memória
#### a) Qual o tamanho em bits do REM, RDM, RI e CI?
#### b) Qual a capacidade da memória, em bytes?
#### c) Se se quisesse aumentar o tamanho das instruções para 17 bits, mantendo inalterado o tamanho do REM, quantas novas instruções poderiam ser criadas?

a) 128 instruções = 2^7, então o código de operação terá 7 bits. A memória tem capacidade de 512 células/palavras, ou seja, 2^9 = 9 bits para o operando. Somando os dois valores, chegamos ao tamanho da instrução, que é 16 bits.

Assim, o REM e o CI terão 16 bits e o RDM e o RI terão 7 bits. 

b) Capacidade da memória em bytes = 512 x 16 = 8192 / 8 = 1024 = 2^10 = 1Kbyte


```python
qtdinst = 128
codigoop = math.log2(qtdinst)
celulas = 512
operando = math.log2(celulas)
instrucao = codigoop + operando
totalbytes = (celulas * instrucao) / 8
print(f"O REM e o CI têm {operando} bits, o RDM e o RI têm {instrucao} bits e a capacidade total da memória em bytes é {totalbytes} (1KB)")
```

c) Se a instrução tivesse 17 bits e o REM continuasse com 9 bits, o código de operação passaria a ter 8 bits, então teríamos 2^8 instruções = 256 instruções. Então poderiam ser criadas 128 novas instruções.
