```python
import math

celulas = 2**4 * 2**10  ## 32KB = Capacidade da MP
bitsendereco = math.log2(celulas)
print(f"A memória principal tem um total de {celulas} bits e o endereço tem {bitsendereco} bits")

bitscelula = 8 # 1 célula = 1 byte = 8 bits
print(f"Cada célula tem um total de {bitscelula} bits")

qtdlinhas = (2**5 * 2**10) / 2**3 ## Capacidade da cache / células por linha
bitslinha = math.log2(qtdlinhas)
print(f"A cache tem um total de {qtdlinhas} linhas e {bitslinha} bits para a linha")

qtdblocos = (2**4 * 2**20) / 2**3 ## Total de células da MP / células por bloco
print(f"A cache tem um total de {qtdblocos} blocos (2^21)") ## 16MB / 8 bytes 

blocoslinha = qtdblocos / qtdlinhas ## Quantidade de blocos / quantidade de linhas
tag = math.log2(blocoslinha)
print(f"Cada linha tem {blocoslinha} blocos e a tag tem {tag} bits de largura")

totalbits = 32000 * 8 + (qtdlinhas * tag) ## Total de bits da cache = tamanho da cache (Capacidade * bits por célula) + (número de linhas * bits de endereço do bloco)
print(f"A cache tem um total de {totalbits} bits")
```

```python
import math

cache = 2**5 * 2**10  ## Capacidade da cache
print(f"A cache tem capacidade de {cache} bits")

celulasmp = 2**4 * 2**20  ## Capacidade da MP
print(f"A MP tem {celulasmp} células (2^24)")

tamanholinha = 8

linhasconjunto = 4

qtdlinhas = cache / tamanholinha ## Capacidade da cache / tamaho da linha
print(f"A cache tem um total de {qtdlinhas} linhas (2^12)")

qtdblocos = celulasmp / 2**3 ## Total de células / células por bloco
print(f"A cache tem um total de {qtdblocos} blocos (2^21)") ## 16MB / 8 bytes 

qtdconjuntos = qtdlinhas / linhasconjunto ## 2**12 / 2**2
print(f"A cache tem um total de {qtdconjuntos} conjuntos (2^10)")

tag = math.log2(qtdblocos / qtdconjuntos)
print(f"A tag tem {tag} bits")

totalbits = cache * 8 + (qtdlinhas * tag) ## Total de bits da cache = tamanho da cache (Capacidade * bits por célula) + (número de linhas * bits para identificar a tag)
print(f"A cache tem um total de {totalbits} bits")
```

## Exemplo 5.9

```python
from IPython.display import Image
Image(filename=r'C:\Users\desk\Documents\exemplos memória cache\exemplo59.png')
```

```python
import math
# 16 bytes = 16 células por bloco 

celulaslinha = 2**4 
bitscelula = math.log2(celulaslinha)
print(f"A memória tem {celulaslinha} células por linha e {bitscelula} bits por célula")

celulas = 2**26 ## 2^6 * 2^20 = 64 MB
bitsendereco = math.log2(celulas)
print(f"A memória tem {celulas} células e cada endereço tem {bitsendereco} bits")

cache = 2**5 * 2**10  ## Capacidade da cache
print(f"A cache tem capacidade de {cache} bits")

qtdblocos = (2**6 * 2**20) / 2**4 ## Células / células por bloco
print(f"A memória tem um total de {qtdblocos} blocos (2^22)") ## 64MB / 16 bytes

qtdlinhas = cache / celulaslinha ## Capacidade da cache / células por linha
print(f"A cache tem um total de {qtdlinhas} linhas (2^11)") 

qtdconjuntos = qtdlinhas / 2**2 ## Quantidade de linhas / linhas por conjunto (4)
bitsconjuntos = math.log2(qtdconjuntos)
print(f"A cache tem um total de {qtdconjuntos} conjuntos e cada conjunto tem {bitsconjuntos} bits")

tag = math.log2(qtdblocos / qtdconjuntos)
print(f"A tag tem {tag} bits")

print(f"O endereço terá {bitsendereco} bits: {bitsconjuntos} bits para os conjuntos, {bitscelula} para a célula e {tag} bits para a tag")
```
