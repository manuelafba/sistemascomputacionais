## Projete um sistema de memória que possui 13 pinos de endereços e um barramento de dados (BD) que transporta 2 palavras por vez. O BD não pode exceder 64 bits e a palavra deve ser maior que a célula da memória. A memória deve ter capacidade de 8 Gbits. Depois do projeto, indique qual é o tamanho da palavra, da célula, do BD e do barramento de endereços, e desenhe o chip.

```python
from IPython.display import Image
Image(filename=r'C:\Users\desk\Pictures\ProjetoChip1.png')
```

Capacidade da memória = 8Gbits

8Gbits = 2^33

Como a questão diz que o BD não pode exceder 64 bits, considerei o máximo possível (64 bits). A palavra vai ter 32bits, já que o BD transporta duas palavras por vez. Também é dito que a célula deve ser menor que a palavra, logo a célula terá 16 bits.

2^33 = tamanho do endereço * tamanho da célula

Tamanho do endereço = 2^33 / 2^4 = 2^29 = 2^20 * 2^9 = 512Mbits

Considerando uma matriz quadrada e os 13 bits de endereço dados na questão, as linhas e as colunas terão o mesmo número de entradas, logo RAS = CAS = 2^13

Teremos assim 2^33 = 2^13 * 2^13 * 2^4

Porém, podemos observar que ainda faltam 3 unidades no expoente para completar 33. Essas 3 unidades são destinadas aos bancos. 

Logo: 2^33 = 2^13 * 2^13 * 2^4 * 2^3

Tamanho da célula: 16bits

Tamanho da palavra: 32bits

Tamanho do barramento de endereço: 19bits

Tamanho do barramento de dados: 64bits

```python
from IPython.display import Image
Image(filename=r'C:\Users\desk\Downloads\ProjetoChip2.png')
```

O chip tem no total 32Gbits

32Gbits = 2^35

A questão diz que a memória deve ter 16 linhas de dados. Considerei que o BD lê uma célula por vez, logo a célula e a palavra terão 16 bits.

Tamanho do endereço = 2^35 / 2^4 = 2^31 = 2^30 * 2^1 = 2Gbits

Os pinos de endereço devem ser menores que 14, considerei o máximo (13) e uma matriz quadrada 13x13

Temos: 2^35 = 2^13 * 2^13 * 2^4

Faltam 5 unidades para completar 35 unidades no expoente, logo teremos 2^5 bancos

Assim: 2^35 = 2^13 * 2^13 * 2^4 * 2^5

Tamanho do barramento de endereçs: 21 bits

Tamanho do barramento de dados: 16 bits

Célula e palavra: 16 bits
